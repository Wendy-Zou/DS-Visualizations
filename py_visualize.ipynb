```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(color_codes = True)
import warnings
warnings.filterwarnings("ignore")
```

```python
salary = pd.read_csv('ds_salaries.csv')
```

```python
salary
```

```python
salary["job_title"].value_counts()
```

```python
salary.plot(kind='scatter', x = 'work_year', y = 'salary_in_usd')
plt.show()
```

```python
sns.jointplot(x = 'employment_type', y = 'salary_in_usd', data = salary, size = 8)
plt.show()
```

```python
sns.FacetGrid(salary, hue='company_size', palette='muted').map(plt.scatter, 'experience_level', 'salary_in_usd').add_legend()
plt.show()
# hue='species' gives each species a different color
```

```python
sns.boxplot(x='company_size', y='salary_in_usd', palette='husl', data=salary)
plt.show()
```

```python
sns.boxplot(x='work_year', y ='salary_in_usd', palette='husl', data = salary)
plt.show()
```

```python
sns.violinplot(x='company_size', y='salary_in_usd', palette='muted', data=salary)
plt.show()
```

Violin plot with the datapoints overlayed on the figures. This gives us a better idea of the data distribution.

```python
#plotting a violin plot
sns.violinplot(x='company_size', y='salary_in_usd', palette='muted', data=salary)

#plotting a swarm plot on top of the violin plot
sns.swarmplot(x='company_size', y='salary_in_usd', data=salary, color='w', alpha=.9);
plt.show()
```

```python
sns.FacetGrid(salary, hue='company_size', palette='husl').map(sns.kdeplot, 'salary_in_usd').add_legend()
plt.show()
```

Pairplot - depicts the bivariate relation between each pair of features.

```python
sns.pairplot(salary, hue='company_size', palette='husl')
plt.show()
```

```python
sns.histplot(salary['experience_level'], kde=True, bins=20)
```

```python
sns.displot(data=salary, x="salary_in_usd", kde=True)
```

```python
# .barplot() displays the mean by default
sns.barplot(x="company_size", y="salary_in_usd", data=salary, palette="deep")
```

```python
sns.stripplot(x='work_year', y='salary_in_usd', data=salary, hue='company_size', palette='husl', dodge=True)
```

```python
# gives correlation coefficients
salary.corr()
```

```python
sns.heatmap(salary.corr(), annot=True, cmap='coolwarm')
```
